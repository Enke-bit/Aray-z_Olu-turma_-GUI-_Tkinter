```python
from tkinter import *
from tkcalendar import DateEntry

# part 1

master =Tk()

# pack
# place
# grid


# --> burada arayüzün ana yapısını oluşturuyorum.
canvas = Canvas(master, height=450, width=750)
# canvas burada çizim yapılabilir ekran anlamına gelmektedir. Burada yapıcağım arayüzün alan büyüklüğünü tanımlıyorum.
canvas.pack()
# bu pack ise yaptığım özeştirmenin çalışması için gerekli.


# --> frame ile ise bu yapını wigitch bölümlerinin conteyner gibi toplanmasını sağlıyor. örneğimdeki mavi arka alan.
frame_ust = Frame(master, bg='#1e81b0')
frame_ust.place(relx=0.1, rely=0.1, relwidth=0.8, relheight=0.1) 
# relx=0.1 --> anlamı x düzleminde yüzde 10'luk bir pay bırak
# rely=0.1 --> anlamı y düzleminde yüzde 10'luk bir pay bırak
# relwidth --> anlamı açılan ekranın yüzde 80 yatay'da al.
# relheight --> anlamı açılan ekranı yüzde 10 dikeyde al.

frame_sol_alt = Frame(master, bg='#1e81b0')
frame_sol_alt.place(relx=0.1, rely=0.21, relwidth=0.23, relheight=0.5) 

frame_sağ_alt = Frame(master, bg='#1e81b0')
frame_sağ_alt.place(relx=0.34, rely=0.21, relwidth=0.56, relheight=0.5) 
# bu oranları deneme ve yanılma olacak şekilde oluyor ama deneyerek daha iyi denmeler yapabilirim.

hatirlatma_tipi_etiket = Label(frame_ust, bg='#1e81b0', text='hatırlatma tipi:', font="Verdana 12 bold")
# --> stil tanımlamalarını bu alana yaparken. Text metin girmemei sağlarken 'font' yazını tipini boyutu gibi özellikleri tanımlıyorum.
hatirlatma_tipi_etiket.pack(padx=10, pady=10, side=LEFT)
# --> Konum tanımlamalarını 'pack()' içerisine yapıyorum.
# --> padx ve pady yazının x ve y düzleminden uzaklığının tanımlıyor.
# --> side ise yazının sağa sola yada ortaya konumunun tanımlanmasını yapmama imkan sağlıyor.

hatirlatma_tipi_opsyon = StringVar(frame_ust) # acilir menü tanımlaması yapıyorum
hatirlatma_tipi_opsyon.set("\t") # defolt ne değer yazacağını tanımlıyorum.

hatirlatma_tipi_acilir_menü = OptionMenu( # menü işlemlerini tanımlıyorum.
    frame_ust,
    hatirlatma_tipi_opsyon,
    "Doğum günü",
    "Evlilik yıl dönümü",
    "Cumhuriyetin ilani") # ebeveyn frame tanımladıktan sonra içerisinde yazılacak olan ifadeleri tanımlıyorum.
hatirlatma_tipi_acilir_menü.pack(padx=10, pady=10, side=LEFT)

hatirlatma_tarih_etiket = Label(frame_ust, bg='#1e81b0', text='hatırlatma tarih:', font="Verdana 12 bold")
hatirlatma_tarih_etiket.pack(padx=10, pady=10, side=LEFT)

hatirlatma_tarih_secici = DateEntry(frame_ust, width=12, background='orange', foreground='black', borderwidth=1, locale="de_DE")
hatirlatma_tarih_secici._top_cal.overrideredirect(False) # tarih verisi gireceğim alana özel bir tanımlama yapmama gerekiyor onun girişini yapıyorum.
hatirlatma_tarih_secici.pack(padx=10, pady=10, side=LEFT)


# Part 2

Label(frame_sol_alt, bg='#1e81b0', text='Hatırlatma yöntemi:', font="Verdana 10 bold").pack(padx=10, pady=10, anchor=NW)
# --> anchor aslında harita mantığı ile kuvey güney doğu batı gibi çalışır. Bu örnekte NW kuzey batı çalışır.

var = IntVar()
# --> Burada ya onu ya onu saçme işlemi yapaçağım için ve bunlar aynı alandaki tercihler olucağı için bu şekilde tanımladım.

R1 = Radiobutton(frame_sol_alt, text="sisteme kaydet", variable = var, value=1, bg='#1e81b0', font="Verdana 8")
# --> 'variable'  ile ilgili varı tanımladım. Value ile de 1 eşitledim.
R1.pack(anchor=NW, pady=5, padx=15)

R2 = Radiobutton(frame_sol_alt, text="e-posta gönder", variable = var, value=2, bg='#1e81b0', font="Verdana 8")
# --> 'variable'  ile ilgili varı tanımladım. Value ile de 2 eşitledim çünkü var için ikinci girdi.
R2.pack(anchor=NW, pady=5, padx=15)

var1 = IntVar() # her checkbox için yeni var tanımlamamın sebebi bunların tek bir anlam ifade etmesi ver birbirlerinden farklı olması.
C1 = Checkbutton(frame_sol_alt, text='bir hafta önce', variable=var1, onvalue=1, offvalue=0, bg='#1e81b0', font="Verdana 8")
C1.pack(anchor=NW, pady=2, padx=25)
# --> onvalue ve offvalue olmasının sebebi eş zamanlı seçilebilir olmasıdır.
var2 = IntVar()
C2 = Checkbutton(frame_sol_alt, text='bir gün önce', variable=var2, onvalue=1, offvalue=0, bg='#1e81b0', font="Verdana 8")
C2.pack(anchor=NW, pady=2, padx=25)

var3 = IntVar()
C3 = Checkbutton(frame_sol_alt, text='ayni gün', variable=var3, onvalue=1, offvalue=0, bg='#1e81b0', font="Verdana 8")
C3.pack(anchor=NW, pady=2, padx=25)


# Part 3
from tkinter import messagebox
# --> mesaj ekranı çıkmasını yarayan bir kütüpane kulanaçağım için bu kütüpaneyi import ediyorum.
def gonder():
    son_mesaj = ''
    try:
        if var.get(): # burdaki ana mantık radio button seçilmişmi onu sorguluyorum.
            if var.get() == 1: # eğer kulanıcı tarafından burdaki variable ile 1 değiri gelmiş mi yani etkileşime birmişmi? eğer girdiyse ilk radio buttondur bu.
                son_mesaj += 'Veriniz başarı ile sisteme kaydedilmiştir.' # etkileşim olduğunda çıkmasını istediğim uygulama mesajı

                # --> Bu alana gerçekten sisteme kaydedilme durumunu gerçekleştireçeğim.
                tip = hatirlatma_tipi_opsyon() if hatirlatma_tipi_opsyon.get()=='' else "Genel"
                # --> acılır menü için veri çekmek. ve yaptığım if döndüsü tek satırda if döngüsüdür.
                tarih = hatirlatma_tarih_secici.get()
                # --> tarih seçiçiden tarih verisini aldım ve bu boş olmaz o yüzden if'li tek satır yapımı yazmadım.
                mesaj = text_alani.get("1.0", "end")
                # --> text verisinde get çalışmadığı için bu şekilde ordan veri getiriyoruz.

                # --> şimdi dosya açıp dosya kapatma işlemlerini yapaçağım.
                with open("Harırlatma.txt", "w") as dosya: # dosya açıtım ve ismini hatırlatma koydum uzantısını verdim, w ise yaz anlamına geliyor.
                    dosya.write( # dosya içini tanımlıyorum nasıl yazacağının tipini
                        '{} katagorisinde, {} tarhinde, ve "{}" notuyla hatirlatma'.format( # tnımladığım tipden verilerin nasıl geleceğini tanımlıyorum
                            tip,
                            tarih,
                            mesaj

                        ))
                    dosya.close()

            elif var.get() == 2:
                son_mesaj += 'Veriniz başarı ile e-posta gönderilmiştir.'

            messagebox.showinfo("başarılı işlem", son_mesaj)
        # --> kulanıcı text verisi girmezse.
        else:
            son_mesaj += 'text verisi girdiğinizden emin olun!' # ekrana bu uyarı yazısıve
            messagebox.showwarning("yetersiz bilgi", son_mesaj) # uyarı penceris açıyorum ekrana.
    # --> genel olarak işlem çalışmazsa yapılacak seneryo için tanımlama yapıyorum
    except:
        son_mesaj += 'işlem başarısız oldu!' 
        messagebox.showerror("başarısız işlem", son_mesaj)
    # her durumda yapmasını istediğim şey ise.
    finally:
        master.destroy()
        



Label(frame_sağ_alt, bg='#1e81b0', text='hatırlatma mesaji:', font="Verdana 10 bold").pack(padx=10, pady=10, anchor=NW)
# --> 3 alanın etiket tanımlamasını yapıtorum.
text_alani = Text(frame_sağ_alt, height=9, width=50)
# --> text alanın kaplıyacağı alanın tanımlamasını yapıyorum.
text_alani.tag_configure('style', foreground='#bfbfbf', font=('Verdana', 7, 'bold'))
# --> text alanın 'tag_configure' kullanarak farklı fiziksel özellikler tanımlıyorum.
text_alani.pack()

karsılama_metni = 'Mesajini buraya gir'
# --> default olarak bir metin tanımlıyorum.
text_alani.insert(END, karsılama_metni, 'style')
# --> bualana betni ve tanıdığım farklı fiziksel alanları insert ediyorum yani ekliyrum.

gonder_butonu = Button(frame_sağ_alt, text='gönder', command=gonder)
# --# buton değişkeni ve onun nerede ne görev yapacağını tanımlıyorum. Ve 'command' ile görev tranımını yazıyorum.
gonder_butonu.pack(anchor=S)
# --> button o frame içindeki konumunu yazıyorum.






master.mainloop()
# mainloop aslında bu kodun sürekli bir loop içerisinde çağırılmasını sağlak ve buda bu hücrenin sürekli çalışmasını sağlar. 
```
