```python
from tkinter import *
from tkcalendar import DateEntry

# part 1

master =Tk()

# pack
# place
# grid


# --> burada arayzn ana yapsn oluturuyorum.
canvas = Canvas(master, height=450, width=750)
# canvas burada izim yaplabilir ekran anlamna gelmektedir. Burada yapcam arayzn alan bykln tanmlyorum.
canvas.pack()
# bu pack ise yaptm zetirmenin almas iin gerekli.


# --> frame ile ise bu yapn wigitch blmlerinin conteyner gibi toplanmasn salyor. rneimdeki mavi arka alan.
frame_ust = Frame(master, bg='#1e81b0')
frame_ust.place(relx=0.1, rely=0.1, relwidth=0.8, relheight=0.1) 
# relx=0.1 --> anlam x dzleminde yzde 10'luk bir pay brak
# rely=0.1 --> anlam y dzleminde yzde 10'luk bir pay brak
# relwidth --> anlam alan ekrann yzde 80 yatay'da al.
# relheight --> anlam alan ekran yzde 10 dikeyde al.

frame_sol_alt = Frame(master, bg='#1e81b0')
frame_sol_alt.place(relx=0.1, rely=0.21, relwidth=0.23, relheight=0.5) 

frame_sa_alt = Frame(master, bg='#1e81b0')
frame_sa_alt.place(relx=0.34, rely=0.21, relwidth=0.56, relheight=0.5) 
# bu oranlar deneme ve yanlma olacak ekilde oluyor ama deneyerek daha iyi denmeler yapabilirim.

hatirlatma_tipi_etiket = Label(frame_ust, bg='#1e81b0', text='hatrlatma tipi:', font="Verdana 12 bold")
# --> stil tanmlamalarn bu alana yaparken. Text metin girmemei salarken 'font' yazn tipini boyutu gibi zellikleri tanmlyorum.
hatirlatma_tipi_etiket.pack(padx=10, pady=10, side=LEFT)
# --> Konum tanmlamalarn 'pack()' ierisine yapyorum.
# --> padx ve pady yaznn x ve y dzleminden uzaklnn tanmlyor.
# --> side ise yaznn saa sola yada ortaya konumunun tanmlanmasn yapmama imkan salyor.

hatirlatma_tipi_opsyon = StringVar(frame_ust) # acilir men tanmlamas yapyorum
hatirlatma_tipi_opsyon.set("\t") # defolt ne deer yazacan tanmlyorum.

hatirlatma_tipi_acilir_men = OptionMenu( # men ilemlerini tanmlyorum.
    frame_ust,
    hatirlatma_tipi_opsyon,
    "Doum gn",
    "Evlilik yl dnm",
    "Cumhuriyetin ilani") # ebeveyn frame tanmladktan sonra ierisinde yazlacak olan ifadeleri tanmlyorum.
hatirlatma_tipi_acilir_men.pack(padx=10, pady=10, side=LEFT)

hatirlatma_tarih_etiket = Label(frame_ust, bg='#1e81b0', text='hatrlatma tarih:', font="Verdana 12 bold")
hatirlatma_tarih_etiket.pack(padx=10, pady=10, side=LEFT)

hatirlatma_tarih_secici = DateEntry(frame_ust, width=12, background='orange', foreground='black', borderwidth=1, locale="de_DE")
hatirlatma_tarih_secici._top_cal.overrideredirect(False) # tarih verisi gireceim alana zel bir tanmlama yapmama gerekiyor onun giriini yapyorum.
hatirlatma_tarih_secici.pack(padx=10, pady=10, side=LEFT)


# Part 2

Label(frame_sol_alt, bg='#1e81b0', text='Hatrlatma yntemi:', font="Verdana 10 bold").pack(padx=10, pady=10, anchor=NW)
# --> anchor aslnda harita mant ile kuvey gney dou bat gibi alr. Bu rnekte NW kuzey bat alr.

var = IntVar()
# --> Burada ya onu ya onu same ilemi yapaam iin ve bunlar ayn alandaki tercihler oluca iin bu ekilde tanmladm.

R1 = Radiobutton(frame_sol_alt, text="sisteme kaydet", variable = var, value=1, bg='#1e81b0', font="Verdana 8")
# --> 'variable'  ile ilgili var tanmladm. Value ile de 1 eitledim.
R1.pack(anchor=NW, pady=5, padx=15)

R2 = Radiobutton(frame_sol_alt, text="e-posta gnder", variable = var, value=2, bg='#1e81b0', font="Verdana 8")
# --> 'variable'  ile ilgili var tanmladm. Value ile de 2 eitledim nk var iin ikinci girdi.
R2.pack(anchor=NW, pady=5, padx=15)

var1 = IntVar() # her checkbox iin yeni var tanmlamamn sebebi bunlarn tek bir anlam ifade etmesi ver birbirlerinden farkl olmas.
C1 = Checkbutton(frame_sol_alt, text='bir hafta nce', variable=var1, onvalue=1, offvalue=0, bg='#1e81b0', font="Verdana 8")
C1.pack(anchor=NW, pady=2, padx=25)
# --> onvalue ve offvalue olmasnn sebebi e zamanl seilebilir olmasdr.
var2 = IntVar()
C2 = Checkbutton(frame_sol_alt, text='bir gn nce', variable=var2, onvalue=1, offvalue=0, bg='#1e81b0', font="Verdana 8")
C2.pack(anchor=NW, pady=2, padx=25)

var3 = IntVar()
C3 = Checkbutton(frame_sol_alt, text='ayni gn', variable=var3, onvalue=1, offvalue=0, bg='#1e81b0', font="Verdana 8")
C3.pack(anchor=NW, pady=2, padx=25)


# Part 3
from tkinter import messagebox
# --> mesaj ekran kmasn yarayan bir ktpane kulanaam iin bu ktpaneyi import ediyorum.
def gonder():
    son_mesaj = ''
    try:
        if var.get(): # burdaki ana mantk radio button seilmimi onu sorguluyorum.
            if var.get() == 1: # eer kulanc tarafndan burdaki variable ile 1 deiri gelmi mi yani etkileime birmimi? eer girdiyse ilk radio buttondur bu.
                son_mesaj += 'Veriniz baar ile sisteme kaydedilmitir.' # etkileim olduunda kmasn istediim uygulama mesaj

                # --> Bu alana gerekten sisteme kaydedilme durumunu gerekletireeim.
                tip = hatirlatma_tipi_opsyon() if hatirlatma_tipi_opsyon.get()=='' else "Genel"
                # --> aclr men iin veri ekmek. ve yaptm if dnds tek satrda if dngsdr.
                tarih = hatirlatma_tarih_secici.get()
                # --> tarih seiiden tarih verisini aldm ve bu bo olmaz o yzden if'li tek satr yapm yazmadm.
                mesaj = text_alani.get("1.0", "end")
                # --> text verisinde get almad iin bu ekilde ordan veri getiriyoruz.

                # --> imdi dosya ap dosya kapatma ilemlerini yapaam.
                with open("Harrlatma.txt", "w") as dosya: # dosya atm ve ismini hatrlatma koydum uzantsn verdim, w ise yaz anlamna geliyor.
                    dosya.write( # dosya iini tanmlyorum nasl yazacann tipini
                        '{} katagorisinde, {} tarhinde, ve "{}" notuyla hatirlatma'.format( # tnmladm tipden verilerin nasl geleceini tanmlyorum
                            tip,
                            tarih,
                            mesaj

                        ))
                    dosya.close()

            elif var.get() == 2:
                son_mesaj += 'Veriniz baar ile e-posta gnderilmitir.'

            messagebox.showinfo("baarl ilem", son_mesaj)
        # --> kulanc text verisi girmezse.
        else:
            son_mesaj += 'text verisi girdiinizden emin olun!' # ekrana bu uyar yazsve
            messagebox.showwarning("yetersiz bilgi", son_mesaj) # uyar penceris ayorum ekrana.
    # --> genel olarak ilem almazsa yaplacak seneryo iin tanmlama yapyorum
    except:
        son_mesaj += 'ilem baarsz oldu!' 
        messagebox.showerror("baarsz ilem", son_mesaj)
    # her durumda yapmasn istediim ey ise.
    finally:
        master.destroy()
        



Label(frame_sa_alt, bg='#1e81b0', text='hatrlatma mesaji:', font="Verdana 10 bold").pack(padx=10, pady=10, anchor=NW)
# --> 3 alann etiket tanmlamasn yaptorum.
text_alani = Text(frame_sa_alt, height=9, width=50)
# --> text alann kaplyaca alann tanmlamasn yapyorum.
text_alani.tag_configure('style', foreground='#bfbfbf', font=('Verdana', 7, 'bold'))
# --> text alann 'tag_configure' kullanarak farkl fiziksel zellikler tanmlyorum.
text_alani.pack()

karslama_metni = 'Mesajini buraya gir'
# --> default olarak bir metin tanmlyorum.
text_alani.insert(END, karslama_metni, 'style')
# --> bualana betni ve tandm farkl fiziksel alanlar insert ediyorum yani ekliyrum.

gonder_butonu = Button(frame_sa_alt, text='gnder', command=gonder)
# --# buton deikeni ve onun nerede ne grev yapacan tanmlyorum. Ve 'command' ile grev tranmn yazyorum.
gonder_butonu.pack(anchor=S)
# --> button o frame iindeki konumunu yazyorum.






master.mainloop()
# mainloop aslnda bu kodun srekli bir loop ierisinde arlmasn salak ve buda bu hcrenin srekli almasn salar. 
```
